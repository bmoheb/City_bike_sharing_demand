# **Bike Sharing Demand Analysis in Washington D.C.**

## **Description:**
Bike sharing systems redefine traditional bike rentals, automating the entire process from membership to rental and return. Users can effortlessly rent bikes from one location and return them to another, adding a new layer of convenience to urban transportation. These systems play a pivotal role in addressing urban challenges by contributing to traffic reduction, offering an alternative mode of transportation to alleviate congestion, especially in urban areas. Additionally, the widespread use of bikes as an eco-friendly mode of transport significantly reduces carbon emissions, fostering cleaner and greener urban environments. Moreover, by encouraging physical activity, bike sharing promotes a healthier lifestyle among users, positively impacting public health.


## **Objectives:**
Conduct a comprehensive analysis of the Washington D.C. Bike Sharing dataset to uncover the correlation between the number of rented bikes and various factors, including temperature, season, and prevailing weather conditions.

## **Dataset:**
*While the dataset includes both daily and hourly data, our analysis specifically utilizes the daily count of rental bikes from 2011 to 2012 in Washington, D.C.'s [Capital Bikeshare system](https://capitalbikeshare.com/system-data) along with corresponding weather and seasonal information.*

### **Content:**

The fields in both hour.csv and day.csv are identical, with the exception of the 'hr' field, which is not present in day.csv.

- **instant:** Record index
- **dteday:** Date
- **season:** Season (1:springer, 2:summer, 3:fall, 4:winter)
- **yr:** Year (0: 2011, 1:2012)
- **mnth:** Month (1 to 12)
- **hr:** Hour (0 to 23)
- **holiday:** Weather day is holiday or not (extracted from Holiday Schedule)
- **weekday:** Day of the week
- **workingday:** If the day is neither a weekend nor a holiday, it is 1; otherwise, it is 0.
- **weathersit:** (Extracted from Freemeteo)
  - 1: Clear, Few clouds, Partly cloudy, Partly cloudy
  - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
  - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
  - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog
- **temp:** Normalized temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-8, t_max=+39 (only in hourly scale)
- **atemp:** Normalized feeling temperature in Celsius. The values are derived via (t-t_min)/(t_max-t_min), t_min=-16, t_max=+50 (only in hourly scale)
- **hum:** Normalized humidity. The values are divided by 100 (max)
- **windspeed:** Normalized wind speed. The values are divided by 67 (max)
- **casual:** Count of casual users
- **registered:** Count of registered users
- **cnt:** Count of total rental bikes, including both casual and registered

# Let's dive into the coding, analysis and data visualization

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Set matplotlib to display plots in the Jupyter Notebooke
%matplotlib inline
```

```python
# Load the dataset
dataset_path = 'day.csv'
bike_sharing_data = pd.read_csv(dataset_path)
# Replace infinite values with NaN
bike_sharing_data.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Explore the dataset
bike_sharing_data.info()
```

```python
# Descriptive data statistics
selected_columns = ['temp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
bike_sharing_data[selected_columns].describe().round(2)
```

### Descriptive statistics simply summarize the main features of the data.

```python
# Identify key columns
key_columns = ['dteday', 'temp', 'weathersit', 'season', 'hum', 'windspeed','cnt', 'casual', 'registered']
bike_sharing_data[key_columns].head()
```

### Identifying key columns in your dataset provides insight and direction for analysis. It helps determine the required data visualizations and guides the features of the dataset to study.

```python
# Convert 'dteday' column to datetime type
bike_sharing_data['dteday'] = pd.to_datetime(bike_sharing_data['dteday'])
```

### Converting the 'dteday' column to datetime format standardizes date representation, enables time-based operations, and facilitates analysis and visualization of temporal data.

```python
# Group by month and calculate the sum of casual, registered, and total rented bikes
monthly_data = bike_sharing_data.groupby(bike_sharing_data['dteday'].dt.to_period("M")).agg({
    'casual': 'sum',
    'registered': 'sum',
    'cnt': 'sum'
})

# Convert the Period index to string with 'YYYY-MM' format
monthly_data.index = monthly_data.index.strftime('%Y-%m')

# Create an array of indices for the bars
indices = np.arange(len(monthly_data.index))

# Set the width of the bars
bar_width = 0.3

# Create a subplot for the monthly breakdown
plt.figure(figsize=(20, 7))

# Plot the monthly total bikes rented along with bar charts for casual and registered users
plt.bar(indices - bar_width, monthly_data['casual'], width=bar_width, color='navy', alpha=0.7, label='Casual Users')
plt.bar(indices, monthly_data['registered'], width=bar_width, color='red', alpha=0.7, label='Registered Users')
plt.plot(indices, monthly_data['cnt'], linestyle='-', color='dimgrey', marker='o', label='Total Bikes Rented')

# Add data labels above the linear graph
for i, value in enumerate(monthly_data['cnt']):
    plt.text(indices[i], value + 50, str(value), ha='center', va='bottom', fontsize=9, color='black')

plt.title('Monthly Breakdown of Bikes Rented')
plt.xticks(indices, monthly_data.index, rotation=45, ha='right')  # Set x-axis labels with 'YYYY-MM' format
plt.xlabel('Month')
plt.ylabel('Number of Users / Bikes Rented')
plt.legend()
plt.grid(True)
plt.show()

print(monthly_data.to_markdown())
```

### It is always a good idea to have a table below your plots for data validation purposes.
### In overall, we see an increase in the total number of rented bikes from 2011 to 2012.

```python
# Define a function to map dates to seasons
def get_season(date):
    if date.month in [3, 4, 5]:
        return 'Spring'
    elif date.month in [6, 7, 8]:
        return 'Summer'
    elif date.month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Winter'

# Create a new column 'new_season' based on the 'dteday' column to avoid conflicts
bike_sharing_data['new_season'] = bike_sharing_data['dteday'].apply(get_season)

# Group by year and season and calculate the sum of casual, registered, and total rented bikes
seasonal_data_new = bike_sharing_data.groupby([bike_sharing_data['dteday'].dt.year, 'new_season']).agg({
    'casual': 'sum',
    'registered': 'sum',
    'cnt': 'sum'
})

# Check if 'new_season' exists in the index
if 'new_season' not in seasonal_data_new.index.names:
    # Convert the index to 'YYYY - Season' format
    seasonal_data_new.index = [f"{year} - {new_season}" for (year, new_season) in seasonal_data_new.index]
    # Set 'new_season' as part of the index
    seasonal_data_new.set_index('new_season', append=True, inplace=True)

# Separate data for each year
seasonal_data_2011 = seasonal_data_new.xs(2011, level=0)
seasonal_data_2012 = seasonal_data_new.xs(2012, level=0)

# Sort seasons within each year
seasonal_data_2011_sorted = seasonal_data_2011.loc[['Spring', 'Summer', 'Fall', 'Winter']]
seasonal_data_2012_sorted = seasonal_data_2012.loc[['Spring', 'Summer', 'Fall', 'Winter']]

# Add a year level to the index
seasonal_data_2011_sorted.index = pd.MultiIndex.from_product([[2011], seasonal_data_2011_sorted.index])
seasonal_data_2012_sorted.index = pd.MultiIndex.from_product([[2012], seasonal_data_2012_sorted.index])

# Concatenate the sorted dataframes back together
seasonal_data_sorted = pd.concat([seasonal_data_2011_sorted, seasonal_data_2012_sorted])

# Extract year and season for x-axis labels
years = seasonal_data_sorted.index.get_level_values(0).astype(str)
seasons = seasonal_data_sorted.index.get_level_values(1)

# Combine year and season for x-axis labels
x_labels = [f"{year} - {season}" for year, season in zip(years, seasons)]

# Create a subplot for the seasonal breakdown
plt.figure(figsize=(20, 7))

# Plot the seasonal total bikes rented along with bar charts for casual and registered users
plt.bar(np.arange(len(seasonal_data_sorted)), seasonal_data_sorted['casual'], width=0.3, color='navy', alpha=0.7, label='Casual Users')
plt.bar(np.arange(len(seasonal_data_sorted)) + 0.3, seasonal_data_sorted['registered'], width=0.3, color='red', alpha=0.7, label='Registered Users')
plt.plot(np.arange(len(seasonal_data_sorted)), seasonal_data_sorted['cnt'], linestyle='-', color='dimgrey', marker='o', label='Total Bikes Rented')

# Add data labels above the linear graph
for i, value in enumerate(seasonal_data_sorted['cnt']):
    plt.text(i, value + 50, str(value), ha='center', va='bottom', fontsize=9, color='black')

plt.title('Seasonal Breakdown of Bikes Rented')
plt.xticks(np.arange(len(seasonal_data_sorted)), x_labels, rotation=45, ha='right')  # Set formatted x-axis labels
plt.xlabel('Year - Season')
plt.ylabel('Number of Users / Bikes Rented')
plt.legend()
plt.grid(True)
plt.show()

print(seasonal_data_sorted.to_markdown())
```

### As anticipated, the peak number of rented bikes occurs during the summer months every year, while the total number of rented bikes reduces during the winter months.

```python
# Define the order of seasons
season_order = ['Spring', 'Summer', 'Fall', 'Winter']

# Create the box plot without specifying colors
plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(x='new_season', y='cnt', data=bike_sharing_data, order=season_order)

# Add title and labels
plt.title('Rented Bikes Distribution in Each Season')
plt.xlabel('Season')
plt.ylabel('Rented Bikes Count')

# Calculate the mean for each season
seasonal_means = bike_sharing_data.groupby('new_season')['cnt'].mean()

# Add data labels for mean
for i, season in enumerate(season_order):
    plt.text(i, seasonal_means[season], f'{seasonal_means[season]:.0f}', color='black', ha='center', va='bottom')

# Show the plot
plt.show()
```

### The box plot visually compares the number of rented bikes across seasons, highlighting the median, quartiles, and outliers. The lower and upper whiskers represent the minimum and maximum numbers of rented bikes in each season.

```python
# Drop the 'new_season' column from the DataFrame
bike_sharing_data_heatmap = bike_sharing_data.drop(['new_season'], axis=1)

# Define the columns to keep in the heatmap
columns_to_keep = ['holiday', 'weekday', 'workingday', 'weathersit', 'temp', 'hum', 'windspeed', 'cnt']

# Keep only the selected columns in the DataFrame
bike_sharing_data_heatmap = bike_sharing_data_heatmap[columns_to_keep]

# Calculate the correlation matrix
correlation_matrix_heatmap = bike_sharing_data_heatmap.corr()

# Create the heatmap with data labels for each pair of variables
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(correlation_matrix_heatmap, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

# Iterate over each cell in the correlation matrix
for i in range(len(correlation_matrix_heatmap)):
    for j in range(len(correlation_matrix_heatmap)):
        # Add the correlation value as text to the cell
        heatmap.text(j + 0.5, i + 0.5, f"{correlation_matrix_heatmap.iloc[i, j]:.2f}",
                     ha='center', va='center', color='black')

plt.title('Correlation Heatmap for Selected Variables')
plt.show()
```

### The heatmap displays correlations between selected variables, with colors indicating the strength and direction: warmer colors for positive correlations and cooler colors for negative correlations.
### It can be seen that temperature and total number of rented bikes are highly correlated.

```python
# Define the order of seasons
season_order = ['Spring', 'Summer', 'Fall', 'Winter']

# Define the order of weather situations and their descriptions
weather_descriptions = {
    1: 'Clear, Few clouds, Partly cloudy, Partly cloudy',
    2: 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist',
    3: 'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds',
    4: 'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'
}

# Group the data by new_season and weathersit and count the instances
weather_counts = bike_sharing_data.groupby(['new_season', 'weathersit']).size().unstack(fill_value=0)

# Add a column for weathersit number 4 with counts set to zero
if 4 not in weather_counts.columns:
    weather_counts[4] = 0

# Reorder the columns to match the weather_order
weather_counts = weather_counts.reindex(columns=weather_descriptions.keys(), fill_value=0)

# Create the bar plot
plt.figure(figsize=(12, 6))

# Create the bar plot for number of instances of each weather situation per season
weather_counts.plot(kind='bar', stacked=False)

# Create the linear chart for total rented bikes per weathersit
try:
    plt.plot(weather_counts.index, bike_sharing_data.groupby('weathersit')['cnt'].sum(), marker='o', color='red', linestyle='-')
except ValueError:
    pass

# Add title and labels
plt.title('Weather Situations by Season and Total Rented Bikes per Weather Situation')
plt.xlabel('')
plt.ylabel('Number of Instances')
plt.legend(title='Weather Situation', loc='upper left')

# Display the table
total_counts = bike_sharing_data.groupby(['new_season', 'weathersit'])['cnt'].sum().reset_index()
total_counts['Description'] = total_counts['weathersit'].map(weather_descriptions)
total_counts = total_counts.pivot_table(index='new_season', columns='weathersit', values='cnt', fill_value=0)
total_counts.columns = total_counts.columns.map(weather_descriptions)
total_counts['Total Count'] = total_counts.sum(axis=1)
total_counts = total_counts.reindex(columns=list(weather_descriptions.values()) + ['Total Count'])

# Sort the seasons chronologically
total_counts = total_counts.reindex(season_order)

# Add a row for total counts of each weather condition
total_counts.loc['Total'] = total_counts.sum()

# Remove the header for season
total_counts.index.name = None

# Replace non-finite values (NaN or inf) with 0
total_counts = total_counts.fillna(0)

# Format total counts as integers
total_counts = total_counts.astype(int)

# Show the plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display the table
display(total_counts)
```

### The bar chart shows that the majority of bike rentals happen in good weather condition where the sky is clear or partly cloudy.
